```python

sim = IndoorORCASim(IndoorOrcaSimConfig())
#Simple example of running simulation with a single agent
#A multi-agent simulation, where all agents go to center

num_agents = 4
agent_positions = [(1,1), (-1,1), (1,-1), (-1,-1)]
#velocities to head at the center
agent_velocities = [(-1.,-1.), (1.,-1.), (-1.,1.), (1.,1.)]
for i in range(num_agents):
    sim.add_agent(agent_positions[i])
    sim.set_agent_pref_velocity(i, agent_velocities[i])


```

```python
for step in range(150):
    sim.do_step()

```
